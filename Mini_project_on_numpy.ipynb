## Loading the Dataset

The dataset is loaded using the NumPy `loadtxt()` function. The CSV file is converted into a NumPy array, allowing efficient numerical computations and data analysis.

Parameters:
- delimiter="," : Specifies that values are separated by commas.
- skiprows=1 : Skips the header row containing column names.

## Understanding Dataset Dimensions

The `shape` attribute is used to determine the number of rows and columns present in the dataset.

Rows represent student records.
Columns represent student attributes such as age, attendance, study hours, and marks.

```python
import numpy as np

data = np.loadtxt(
    "student_dataset.csv",
    delimiter=",",
    skiprows=1
)

print(data.shape)
```

## Statistical Analysis of Marks

Statistical functions are used to understand the overall performance of students.

The following metrics are calculated:
- Average Marks using `mean()`
- Highest Marks using `max()`
- Lowest Marks using `min()`

These metrics help summarize student performance.

```python
print(np.mean(data[:, 4]))
```

```python
print(np.max(data[:, 4]))
```

```python
print(np.min(data[:, 4]))
```

## Identifying High-Performing Students

Boolean indexing is used to filter students who scored more than 90 marks.

This helps identify top-performing students in the dataset.

```python
top_students = data[data[:, 4] > 90]
print(len(top_students))
```

## Attendance Analysis

Students having attendance below 75% are identified using filtering operations.

Low attendance may indicate students who are at risk of poor academic performance.

```python
low_attendance = data[data[:, 2] < 75]
print(len(low_attendance))
```

```python
import numpy as np

data = np.loadtxt(
    "student_dataset.csv",
    delimiter=",",
    skiprows=1
)
```

## Total Number of Students

The total number of student records is calculated using the `len()` function.

This provides the overall size of the dataset.

```python
print("Total Students:", len(data))
```

## Average Age and Attendance

The average age and attendance percentage of students are calculated using NumPy aggregate functions.

These statistics provide insights into the general characteristics of the student population.

```python
print("Average Age:", np.mean(data[:,1]))
```

```python
print("Average Attendance:", np.mean(data[:,2]))
```

## Students Scoring Above 80 Marks

Students who scored above 80 marks are filtered from the dataset.

This analysis helps determine the number of students with strong academic performance.

```python
above_80 = data[data[:,4] > 80]

print("Students Scoring Above 80:", len(above_80))
```

```python
below_40 = data[data[:,4] < 40]

print("Students Scoring Below 40:", len(below_40))
```

## Top 10 Marks

The marks column is sorted in ascending order, and the highest ten marks are selected.

This analysis highlights the best academic performances in the dataset.

```python
top_10 = np.sort(data[:,4])[-10:]

print("Top 10 Marks:")
print(top_10)
```

## Pass Percentage Analysis

Students scoring 40 marks or above are considered passed.

The pass percentage is calculated using:

Pass Percentage = (Number of Passed Students / Total Students) × 100

This metric provides an overview of the overall success rate of students.

```python
passed = data[data[:,4] >= 40]

pass_percentage = (len(passed) / len(data)) * 100

print("Pass Percentage:", pass_percentage)
```

## Additional Performance Metrics

Further analysis includes:

- Highest Mark
- Lowest Mark
- Average Study Hours
- Students with Attendance Below 75%
- Students Studying More Than 8 Hours Per Day

These metrics provide deeper insights into student behavior and academic performance.

```python
# Highest mark
print(np.max(data[:,4]))

# Lowest mark
print(np.min(data[:,4]))

# Average study hours
print(np.mean(data[:,3]))

# Students with attendance below 75%
print(len(data[data[:,2] < 75]))

# Students studying more than 8 hours
print(len(data[data[:,3] > 8]))
```
